```python
import pandas as pd
import numpy as np

df = pd.read_csv('Student Depression Dataset.csv')

df.head()
```

```python
df.info()
df.describe()
```

```python
# Pisahkan kolom numerik dan kategorikal
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Isi numerik dengan mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Isi kategorikal dengan modus
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.lower()
```

```python
from sklearn.preprocessing import LabelEncoder

# Redefine cat_cols after column names have been lowercased
cat_cols = df.select_dtypes(include=['object']).columns

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
if 'study_hours' in df.columns and 'sleep_hours' in df.columns:
    df['study_sleep_ratio'] = df['study_hours'] / (df['sleep_hours'] + 1)
```

```python
df = df.drop_duplicates()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

num_cols = df.select_dtypes(include=['int64','float64']).columns

df[num_cols] = scaler.fit_transform(df[num_cols])
```
